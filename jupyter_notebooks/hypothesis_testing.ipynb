```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# I load my feature-engineered dataset so I have all the human-readable categorical columns
df = pd.read_csv("../data/feature_engineered.csv")

sns.set_theme(style="whitegrid")
```

## 1. Job type vs. term deposit subscription
My hypothesis:

Null (H₀): Job type and term deposit subscription are independent (no relationship).

Alternative (H₁): Job type and term deposit subscription are dependent (there is a relationship).

```python
# Chi-square test
job_table = pd.crosstab(df['job'], df['deposit'])
chi2, p, dof, expected = stats.chi2_contingency(job_table)

print(f"Chi2 Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("I reject the null hypothesis — job type and subscription are related.")
else:
    print("I fail to reject the null hypothesis — no significant relationship found.")

# Visualisation
plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x='job',
    y=df['deposit'].map({'yes': 1, 'no': 0}),
    order=df['job'].value_counts().index,
    palette='viridis'
)
plt.xticks(rotation=45)
plt.title('Subscription Rate by Job Type')
plt.ylabel('Proportion Subscribed')
plt.xlabel('Job Type')
plt.show()
```

My interpretation:
If the p-value is below 0.05, I conclude that job type significantly affects subscription rates. This means certain job categories are more or less likely to subscribe.

## 2. Age group vs. term deposit subscription
My hypothesis:

Null (H₀): Age group and term deposit subscription are independent.

Alternative (H₁): Age group and term deposit subscription are dependent.

```python
# Chi-square test
age_table = pd.crosstab(df['age_band'], df['deposit'])
chi2, p, dof, expected = stats.chi2_contingency(age_table)

print(f"Chi2 Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("I reject the null hypothesis — age group and subscription are related.")
else:
    print("I fail to reject the null hypothesis — no significant relationship found.")

# Visualisation
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x='age_band',
    y=df['deposit'].map({'yes': 1, 'no': 0}),
    order=["<25", "25-34", "35-49", "50-64", "65+"],
    palette='coolwarm'
)
plt.title('Subscription Rate by Age Group')
plt.ylabel('Proportion Subscribed')
plt.xlabel('Age Group')
plt.show()
```

My interpretation:
If the p-value is significant, age groups influence subscription rates. This insight could guide marketing strategies targeted at specific age ranges.

## 3. Contact method vs. term deposit subscription
My hypothesis:

Null (H₀): Contact method and term deposit subscription are independent.

Alternative (H₁): Contact method and term deposit subscription are dependent.

```python
# Chi-square test
contact_table = pd.crosstab(df['contact'], df['deposit'])
chi2, p, dof, expected = stats.chi2_contingency(contact_table)

print(f"Chi2 Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("I reject the null hypothesis — contact method and subscription are related.")
else:
    print("I fail to reject the null hypothesis — no significant relationship found.")

# Visualisation
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x='contact',
    y=df['deposit'].map({'yes': 1, 'no': 0}),
    palette='magma'
)
plt.title('Subscription Rate by Contact Method')
plt.ylabel('Proportion Subscribed')
plt.xlabel('Contact Method')
plt.show()
```

My interpretation:
If the p-value is below 0.05, the method used to contact clients plays a role in subscription outcomes. This could mean certain contact channels are more effective.

## Hypothesis Testing Summary
Chi-square tests revealed significant relationships (p < 0.05) between job type, age group, and contact method with term deposit subscription rates. Retired and student customers, middle-aged and older groups, and those contacted via cellular had higher subscription rates.

Insights:

Prioritise high-performing occupations and age segments.

Focus on cellular outreach for better conversion.

Tailor strategies to improve uptake in lower-performing segments.
